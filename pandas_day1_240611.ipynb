# 4. 데이터 병합
- pd.concat : 단순연결
- pd.merge : 특정 key로 결합

## 4-1. pd.concat

```python
df1 = pd.DataFrame({'a': [1, 2, 3, 4], 'b': [0, 0, 0, 0]})
df2 = pd.DataFrame({'a': [1, 2], 'b': [3, 4], 'c': [5, 6]})
df3 = pd.DataFrame({'d': ['test1', 'test2', 'test3'], 'e': ['test1', 'test2', 'test3']})
display(df1, df2, df3)
```

```python
pd.concat([df1, df3], ignore_index=True)
```

## 4-2. pd.merge

```python
df1 = pd.DataFrame({'ID': [1, 2, 3, 4],
                    'Name': ['Alice', 'Bob', 'Charlie', 'David']})

df2 = pd.DataFrame({'아이디': [3, 4, 5, 6],
                    'Age': [25, 30, 35, 40]})

display(df1, df2)
```

## {'left', 'right', 'outer', 'inner'}

```python
pd.merge(df1, df2, left_on='ID', right_on='아이디', how='outer')
```
